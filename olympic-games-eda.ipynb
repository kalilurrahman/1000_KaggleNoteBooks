## Merge Medals and Results

```python
#Genders by Games
games = list(olympics_hosts[olympics_hosts['game_season'] == 'Summer']['game_slug'].unique())
#games.sort()
males = []
females = []
for i in games:
    m = olympics_medals[olympics_medals['slug_game'] == i]
    male = len(m[m.event_gender == 'Men'])
    males.append(male)
    female = len(m[m.event_gender == 'Women'])
    females.append(female)

# visualization
fig,(ax1, ax2) = plt.subplots(1,2, figsize=(20,15))
sns.barplot(x=males,y=games,color='Fuchsia',alpha = 0.8,label='Male', ax=ax1)
sns.barplot(x=females,y=games,color='Red',alpha = 0.8,label='Female', ax=ax1)
ax1.legend(loc='lower right',frameon = False)
ax1.set(xlabel='Gender', ylabel='Games',title = "Summer Olympic Games")
#plt.show()

games = list(olympics_hosts[olympics_hosts['game_season'] == 'Winter']['game_slug'].unique())
#games.sort()
males = []
females = []
for i in games:
    m = olympics_medals[olympics_medals['slug_game'] == i]
    male = len(m[m.event_gender == 'Men'])
    males.append(male)
    female = len(m[m.event_gender == 'Women'])
    females.append(female)

# visualization
sns.barplot(x=males,y=games,color='blue',alpha = 0.8,label='Male', ax=ax2)
sns.barplot(x=females,y=games,color='lime',alpha = 0.8,label='Female', ax=ax2)
#ax2=ax.twinx()
ax2.legend(loc='lower left',frameon = False)
ax2.set(xlabel='Gender', ylabel='Games',title = "Winter Olympic Games")
ax2.set_xlim(ax2.get_xlim()[1], ax2.get_xlim()[0])
ax2.yaxis.set_label_position('right')
ax2.yaxis.set_ticks_position('right')
fig.tight_layout()
plt.title("Gender Proportion")
plt.show()
```

```python
#Merge olympics_hosts and olympics_medals
olympics_hosts.head()
olympics_medals.head()

olympics_results = pd.merge(olympics_hosts, olympics_medals,
                           left_on=  ['game_slug'],
                           right_on= ['slug_game'],
                           how = 'left')
olympics_results.head(10)
```

```python
goldMedals = olympics_results[ (olympics_results['medal_type'] == 'GOLD') & (olympics_results['game_season'] == 'Summer')]
totalGoldMedals = goldMedals.country_3_letter_code.value_counts().reset_index(name='medal_type').head(10)
g = sns.catplot(x="index", y="medal_type", data=totalGoldMedals,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 10 countries")
g.set_ylabels("Number of GOLD Medals")
plt.title('GOLD Medals per Country')


silverMedals = olympics_results[ (olympics_results['medal_type'] == 'SILVER') & (olympics_results['game_season'] == 'Summer')]
totalSilverMedals = silverMedals.country_3_letter_code.value_counts().reset_index(name='medal_type').head(10)
g = sns.catplot(x="index", y="medal_type", data=totalSilverMedals,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 10 countries")
g.set_ylabels("Number of SILVER Medals")
plt.title('SILVER Medals per Country')


bronzeMedals = olympics_results[ (olympics_results['medal_type'] == 'BRONZE') & (olympics_results['game_season'] == 'Summer')]
totalBronzeMedals = bronzeMedals.country_3_letter_code.value_counts().reset_index(name='medal_type').head(10)
g = sns.catplot(x="index", y="medal_type", data=totalBronzeMedals,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 10 countries")
g.set_ylabels("Number of BRONZE Medals")
plt.title('BRONZE Medals per Country')
```

```python
totalMedals_C = olympics_results[ ((olympics_results['medal_type'] == 'BRONZE') | 
                                 (olympics_results['medal_type'] == 'SILVER') |
                                 (olympics_results['medal_type'] == 'GOLD') ) & (olympics_results['game_season'] == 'Summer')]
totalMedals = totalMedals_C.country_3_letter_code.value_counts().reset_index(name='medal_type').head(10)
print(totalMedals.describe())
print(totalMedals.head())
g = sns.catplot(x="index", y="medal_type", data=totalMedals,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 10 countries")
g.set_ylabels("Number of Medals")
plt.title('Medals per Country')
```

```python
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
tmp = olympics_results.groupby(['discipline_title', 'medal_type'])['country_3_letter_code'].agg('count').dropna()
df = pd.DataFrame(tmp).reset_index()
dfG = df[df['medal_type']=='GOLD']
dfS = df[df['medal_type']=='SILVER']
dfB = df[df['medal_type']=='BRONZE']
```

```python
traceG = go.Bar(
    x = dfG['discipline_title'],y = dfG['country_3_letter_code'],
    name="GOLD",
     marker=dict(
                color='gold',
                line=dict(
                    color='yellow',
                    width=1),
                opacity=0.5,
            ),
    text = dfG['discipline_title'],
    #orientation = 'h'
)
traceS = go.Bar(
    x = dfS['discipline_title'],y = dfS['country_3_letter_code'],
    name="SILVER",
    marker=dict(
                color='silver',
                line=dict(
                    color='grey',
                    width=1),
                opacity=0.5,
            ),
    text=dfS['discipline_title'],
    #orientation = 'h'
)
traceB = go.Bar(
    x = dfB['discipline_title'],y = dfB['country_3_letter_code'],
    name="BRONZE",
    marker=dict(
                color='Brown',
                line=dict(
                    color='black',
                    width=1),
                opacity=0.5,
            ),
    text=dfB['discipline_title'],
   # orientation = 'h'
)
data = [traceG, traceS, traceB]
layout = dict(title = 'Medals per sport',
          xaxis = dict(title = 'Sport', showticklabels=True, tickangle=45,
            tickfont=dict(
                size=8,
                color='black'),), 
          yaxis = dict(title = 'Number of medals'),
          hovermode = 'closest',
          barmode='stack',
          showlegend=False,
          width=900,
          height=600,
         )
fig = dict(data=data, layout=layout)
iplot(fig, filename='events-sports1')
```
